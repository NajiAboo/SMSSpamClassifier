```python
dataset = pd.read_csv("../dataset/spam.csv",encoding='ISO-8859-1', usecols=['v1','v2'])
```

```python
import re
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
stopwords = stopwords.words('english')
```

```python
stopwords_custom = [word for word in stopwords if "n\'t" not in word if 'not' not in word if 'n\'' not in word]
```

```python
corpus = []
for i in range(0,len(dataset)):
     review = re.sub('^[A-Za-z0-9]',' ', dataset["messages"][i])
     review = review.lower()
     review = review.split()
     cp = [ porter_stemmer.stem(word, to_lowercase=True) for word in review if word not in stopwords_custom]
     cp = ' '.join(cp)
     corpus.append(cp)
```

```python
corpus
```

```python
y = pd.get_dummies(dataset["labels"], drop_first=True)
```

```python
y
```

```python
#split to train test 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus,y,test_size=0.2) 
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X_train = cv.fit_transform(X_train).toarray()
```

```python
X_train
```

```python
X_test = cv.transform(X_test).toarray()
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
y_pre = rfc.fit(X_train, y_train)
```

```python
pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
